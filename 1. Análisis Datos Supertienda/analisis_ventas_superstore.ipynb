**3. Análisis y Visualización**

```python
# Eliminación de columnas para el análisis
df_analisis = df.drop(['Row ID+O6G3A1:R6','Order ID','Country', 'Ship Date', 'Ship Mode', 'Customer ID','Product ID', 'Profit'], axis=1)
df_analisis.head(10)
```

```python
# Descripción general de los datos
df_analisis.describe()
```

```python
# Ciudades con mayores ventas
ciudades = df_analisis.groupby('City')['Sales'].sum().sort_values(ascending=False)

otras_ciudades = ciudades.iloc[5:].sum()
union_ciudades = pd.concat([ciudades.head(5), pd.Series(otras_ciudades, index=['Otras Ciudades'])])

# Cantidad de ventas por ciudad
ciudades_cantidad = df_analisis.groupby('City')['Sales'].count().sort_values(ascending=False)

otras_ciudades_cantidad = ciudades_cantidad.iloc[5:].sum()
union_ciudades_cantidad = pd.concat([ciudades_cantidad.head(5), pd.Series(otras_ciudades_cantidad, index=['Otras Ciudades'])])

print('*Ventas por ciudades:\n',union_ciudades)
print('*Cantidad de ventas por ciudades:\n',union_ciudades_cantidad)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Primer gráfico de barras
axs[0, 0].bar(ciudades.head(5).index, ciudades.head(5).values, width = 0.5, color = 'orange')
axs[0, 0].set_title('Ciudades con mayores ventas')
axs[0, 0].set_xlabel('Ciudades')
axs[0, 0].set_ylabel('Total ventas')

# Primer gráfico de pastel
axs[0, 1].pie(union_ciudades.values, labels = union_ciudades.index, autopct = '%1.2f%%', startangle = 90)
axs[0, 1].set_title('Ciudades con mayores ventas (%)')
axs[0, 1].legend(title = 'Ciudades', loc = 'right', bbox_to_anchor = (1.5, 0.6))

# Segundo gráfico de barras
axs[1, 0].bar(ciudades_cantidad.head(5).index, ciudades_cantidad.head(5).values, width = 0.5, color = 'orange')
axs[1, 0].set_title('Ciudades con mayor cantidad de ventas')
axs[1, 0].set_xlabel('Ciudades')
axs[1, 0].set_ylabel('Cantidad ventas')

# Segundo gráfico de pastel
axs[1, 1].pie(union_ciudades_cantidad.values, labels = union_ciudades_cantidad.index, autopct = '%1.2f%%', startangle = 90)
axs[1, 1].set_title('Ciudades con mayor cantidad de ventas (%)')
axs[1, 1].legend(title = 'Ciudades', loc = 'right', bbox_to_anchor = (1.5, 0.6))

plt.tight_layout()
plt.show()
```

```python
# Estados con mayores ventas
estados = df_analisis.groupby('State')['Sales'].sum().sort_values(ascending=False)

otros_estados = estados.iloc[5:].sum()
union_estados = pd.concat([estados.head(5), pd.Series(otros_estados, index=['Otros Estados'])])

# Cantidad de ventas por estados
estados_cantidad = df_analisis.groupby('State')['Sales'].count().sort_values(ascending=False)

otras_estados_cantidad = estados_cantidad.iloc[5:].sum()
union_estados_cantidad = pd.concat([estados_cantidad.head(5), pd.Series(otras_estados_cantidad, index=['Otros Estados'])])

print('*Ventas por estados:\n',union_estados)
print('*Cantidad de ventas por estados:\n',union_estados_cantidad)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Primer gráfico de barras
axs[0, 0].bar(estados.head(5).index, estados.head(5).values, width = 0.5, color = 'black')
axs[0, 0].set_title('Estados con mayores ventas')
axs[0, 0].set_xlabel('Estados')
axs[0, 0].set_ylabel('Total ventas')

# Primer gráfico de pastel
axs[0, 1].pie(union_estados.values, labels = union_estados.index, autopct = '%1.2f%%')
axs[0, 1].set_title('Estados con mayores ventas (%)')
axs[0, 1].legend(title = 'Estados', loc = 'right', bbox_to_anchor = (1.7, 0.6))

# Segundo gráfico de barras
axs[1, 0].bar(estados_cantidad.head(5).index, estados_cantidad.head(5).values, width = 0.5, color = 'black')
axs[1, 0].set_title('Estados con mayor cantidad de ventas')
axs[1, 0].set_xlabel('Estados')
axs[1, 0].set_ylabel('Cantidad ventas')

# Segundo gráfico de pastel
axs[1, 1].pie(union_estados_cantidad.values, labels = union_estados_cantidad.index, autopct = '%1.2f%%')
axs[1, 1].set_title('Estados con mayor cantidad de ventas (%)')
axs[1, 1].legend(title = 'Estados', loc = 'right', bbox_to_anchor = (1.7, 0.6))

plt.tight_layout()
plt.show()
```
