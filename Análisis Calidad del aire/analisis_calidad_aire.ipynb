```python
bar_posicion = np.arange(len(Contaminantes))
plt.figure(figsize = (4,3))
plt.bar(bar_posicion - 0.2, Concentracion, width = 0.4, color = ['blue', 'red', 'green', 'yellow'], label = ciudades )
plt.bar(bar_posicion + 0.2, cpm_concentraciones, width = 0.4, color = 'black', label = cpm_nombre)
plt.xticks(bar_posicion, Contaminantes)
plt.title('Ciudades con concentraciones máximas por contaminantes vs Ciudad de mayor población(New York)')
plt.xlabel('Contaminantes')
plt.ylabel('Concentración (ppm)')
plt.legend(title = 'Ciudades')
plt.show()

ciudades_graf = [er_ciudad_co_max, ciudad_NO2_max, ciudad_O3_max, ciudad_SO2_max, cpm_nombre] 
Poblacion_graf = [er_poblacion_co_max, poblacion_NO2_max, poblacion_O3_max, poblacion_SO2_max, cpm_poblacion]

plt.figure(figsize = (5,3))
plt.bar(ciudades_graf, Poblacion_graf, width = 0.5, color = ['blue', 'red', 'green', 'yellow', 'black'])
plt.title('Población de ciudades con concentraciones máximas por contaminantes vs Ciudad de mayor población(New York)')
plt.xlabel('Ciudades')
plt.ylabel('Población')
plt.show()
```

**Contaminación del aire en las ciudades vs población.**

Las ciudades pueden tener diferentes fuentes de contaminación. Por ejemplo, una ciudad con una gran cantidad de tráfico o industrias pesadas puede tener altos niveles de contaminantes como el monóxido de carbono (CO) y el dióxido de nitrógeno (NO2), independientemente de su población. Por otro lado, una ciudad con una gran población pero con políticas estrictas de control de la contaminación y una alta proporción de áreas verdes puede tener niveles más bajos de contaminación. Por ejemplo, Nueva York tiene una población significativamente mayor que Whittier, Alameda, Gainesville y Syracuse, pero tiene niveles más bajos de CO, NO2, O3 y SO2.

Cabe resaltar que las ciudades con una infraestructura más moderna y eficiente pueden tener niveles más bajos de contaminación. Por ejemplo, las ciudades que han invertido en transporte público eficiente y en energías renovables pueden tener niveles más bajos de contaminación, independientemente de su población.
