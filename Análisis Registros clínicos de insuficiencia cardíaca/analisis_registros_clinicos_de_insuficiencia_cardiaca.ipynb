**Creatinine Phosphokinase (CPK):**
La creatinine phosphokinase (CPK) o reatina cinasa, es una enzima que se encuentra en varias partes del cuerpo, incluyendo el corazón, el cerebro y los músculos esqueléticos. Cuando los niveles de CPK son muy altos, generalmente significa que ha habido una lesión o estrés en el tejido muscular, el corazón o el cerebro. Las condiciones que pueden causar niveles altos de CPK incluyen ataques al corazón, distrofia muscular, lesión renal aguda etc.

```python
#Descripción general 

df.creatinine_phosphokinase.describe()
```

```python
#Visualización de valores atípicos mediante diagrama de caja y bigotes.

plt.boxplot(df.creatinine_phosphokinase, labels = ['Creatinine Phosphokinase'])
plt.title('Valores Atípicos')
plt.show()
```

```python
#Limpieza mediante análisis estadístico.

q1 = df.creatinine_phosphokinase.quantile(0.25)
q3 = df.creatinine_phosphokinase.quantile(0.75)

iqr = q3 - q1

li = q1 - 1.5*iqr
ls = q3 + 1.5*iqr

valores_atipicos = df[(df.creatinine_phosphokinase < li) | (df.creatinine_phosphokinase > ls)] 
porcentaje = len(valores_atipicos)/len(df)*100

print("Cantidad de valores atípicos:", len(valores_atipicos))
print('Porcentaje de valores atípicos:', porcentaje)
print('Descripción general de los valores atípico:\n',valores_atipicos.creatinine_phosphokinase.describe())
```

En el caso de los datos clínicos, los valores atípicos pueden ser tanto errores de medición como indicadores de condiciones médicas reales que son importantes para el análisis.Observando las estadísticas, el valor máximo de creatina cinasa es 7861, que es significativamente mayor que el tercer cuartil (Q3 o 75%) de 582. **Ejection Fraction (EF):** 
Es una medida de la cantidad de sangre que se bombea fuera de los ventrículos, o cámaras inferiores del corazón, cada vez que este se contrae. Es el porcentaje de sangre que sale del ventrículo izquierdo cuando el corazón se contrae.

En un corazón sano, la fracción de eyección es del 50% al 70%. Esto significa que con cada latido, del 50% al 70% de la sangre en el ventrículo izquierdo se bombea hacia el cuerpo. Una fracción de eyección baja puede ser un indicador de insuficiencia cardíaca.

```python
#Descripción general 

df.ejection_fraction.describe()
```

```python
#Visualización de valores atípicos mediante diagrama de caja y bigotes.

plt.boxplot(df.ejection_fraction, labels = ['Ejection Fraction'])
plt.title('Valores Atípicos')
plt.show()
```

**Platelets:**
También conocidos como plaquetas, son una componente de la sangre cuya función es reaccionar ante una lesión en los vasos sanguíneos agrupándose, lo que inicia la formación de un coágulo sanguíneo.

Un recuento de plaquetas normal oscila entre 150,000 y 450,000 plaquetas por microlitro de sangre. Tener más de 450,000 plaquetas es una condición llamada trombocitosis; tener menos de 150,000 se conoce como trombocitopenia.

**Serum Creatinine:**
La creatinina sérica es un producto de desecho en la sangre que proviene de la actividad muscular. Normalmente, los riñones eliminan la creatinina de la sangre, pero cuando la función renal se ralentiza, el nivel de creatinina aumenta. El nivel de creatinina sérica se basa en un análisis de sangre que mide la cantidad de creatinina en la sangre y nos indica qué tan bien están funcionando los riñones. Cuando los riñones no funcionan bien, el nivel de creatinina sérica aumenta.

**Serum Sodium:**
La prueba de sodio en sangre, es una prueba de rutina que mide la cantidad de sodio en su sangre. El sodio es un mineral esencial para su cuerpo, particularmente importante para la función nerviosa y muscular.

Los resultados normales para esta prueba son de 135 a 145 mEq/L (miliequivalentes por litro). Un nivel de sodio en sangre inferior a 135 mEq/L se llama hiponatremia.

Basado en los datos proporcionados, se puede observar que la mayoría de los participantes en el estudio de enfermedades cardiovasculares son hombres, representando un 64.88% de la muestra total. Las mujeres, por otro lado, representan el 35.12% restante. Este desequilibrio en la distribución de género puede tener implicaciones en los resultados del estudio. Por ejemplo, si las enfermedades cardiovasculares se manifiestan de manera diferente en hombres y mujeres, los resultados podrían estar sesgados hacia los síntomas y resultados más comunes en los hombres.
Además, si el objetivo del estudio es hacer inferencias sobre la población general, la sobrerrepresentación de hombres podría limitar la aplicabilidad de los resultados a las mujeres.

En conclusión, aunque este estudio proporciona información valiosa sobre las enfermedades cardiovasculares, es importante tener en cuenta la distribución de género de los participantes al interpretar los resultados. Para futuros estudios, sería beneficioso intentar obtener una muestra más equilibrada de hombres y mujeres para asegurar que los resultados sean representativos de toda la población.

Según los datos proporcionados, de los 299 pacientes en el estudio, 96 murieron (32.11%) y 203 sobrevivieron (67.89%). Al desglosar las muertes por género, encontramos que de los 96 pacientes que murieron, 62 eran hombres (64.58%) y 34 eran mujeres (35.42%).

Esto sugiere que, aunque los hombres representaban una mayor proporción de la muestra total (64.88%), la proporción de hombres que murieron (64.58%) es casi la misma. Esto podría indicar que el género no tuvo un impacto significativo en la tasa de mortalidad en este estudio. Por otro lado, aunque las mujeres representaban una menor proporción de la muestra total (35.12%), la proporción de mujeres que murieron (35.42%) es ligeramente mayor. Esto podría sugerir que las mujeres en este estudio tenían un riesgo ligeramente mayor de muerte, aunque la diferencia es pequeña y podría deberse al azar.

En conclusión, aunque los hombres estaban sobrerrepresentados en el estudio, las tasas de mortalidad parecen ser similares entre hombres y mujeres. Sin embargo, sería útil realizar más investigaciones para confirmar estos hallazgos y explorar más a fondo las diferencias de género en las enfermedades cardiovasculares.

```python
muertos_anemia = df[(df.DEATH_EVENT == 1) & (df.anaemia == 1)]['anaemia'].count()
vivos_anemia = df[(df.DEATH_EVENT == 0) & (df.anaemia == 1)]['anaemia'].count()

muertos_diabetes = df[(df.DEATH_EVENT == 1) & (df.diabetes == 1)]['diabetes'].count()
vivos_diabetes = df[(df.DEATH_EVENT == 0) & (df.diabetes == 1)]['diabetes'].count()

muertos_paa = df[(df.DEATH_EVENT == 1) & (df.high_blood_pressure == 1)]['high_blood_pressure'].count()
vivos_paa = df[(df.DEATH_EVENT == 0) & (df.high_blood_pressure == 1)]['high_blood_pressure'].count()

muertos_fumadores = df[(df.DEATH_EVENT == 1) & (df.smoking == 1)]['smoking'].count()
vivos_fumadores = df[(df.DEATH_EVENT == 0) & (df.smoking == 1)]['smoking'].count()


pacientes = ['Anemia', 'Diabetes', 'Presión Alta', 'Fumadores']
pacientes_muertos = [muertos_anemia, muertos_diabetes, muertos_paa, muertos_fumadores]
pacientes_vivos = [vivos_anemia, vivos_diabetes, vivos_paa, vivos_fumadores]

bar_posicion = np.arange(len(pacientes))

plt.bar(bar_posicion - 0.2, pacientes_muertos, width = 0.4, label='Muertos')
plt.bar(bar_posicion + 0.2, pacientes_vivos, width = 0.4, label='Vivos')
plt.xticks(bar_posicion, pacientes)
plt.title('Cantidad de pacientes por condición')
plt.legend(title = 'Estado')
plt.show()

plt.pie(pacientes_muertos, labels = pacientes, autopct = '%1.2f%%', startangle = 90)
plt.title('Pacientes muertos por enfermedades (%)')
plt.legend(title = 'Condición', loc = 'right', bbox_to_anchor = (1.5, 0.5))
plt.show()

plt.pie(pacientes_vivos, labels = pacientes, autopct = '%1.2f%%', startangle = 90)
plt.title('Pacientes vivos por enfermedades (%)')
plt.legend(title = 'Condición', loc = 'right', bbox_to_anchor = (1.5, 0.5))
plt.show()
```
